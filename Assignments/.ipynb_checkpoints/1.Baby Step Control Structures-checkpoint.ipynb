```python
BMI_Index=int(input("Enter the BMI Index :"))
if(BMI_Index<18.5):
    print("Underweight")
elif(BMI_Index>18.5 and BMI_Index<24.9):
    print("Normal Weight")
elif(BMI_Index>25.0 and BMI_Index<29.9):
    print("Over Weight")
elif(BMI_Index>30.0 and BMI_Index<34.9):
    print("Very Over Weight")
elif(BMI_Index>35.0 and BMI_Index<39.9):
    print("Class II Obesity")
else:
    print("Class III Obesity")
              
```
