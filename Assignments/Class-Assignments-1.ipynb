```python
class OddEven():
    
    def OddEven():
        number=int(input("Enter a number: "))
        if(number%2==0):
            result= str(number)+" is Even number"
        else:
            result= str(number)+" is Odd number"
        print(result)
        return result
Result=OddEven.OddEven()
```
